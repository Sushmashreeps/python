**2. Queue:**
    
It is a data structure which uses FIFO rule


**FIFO-First_In_First_Out**

In this rule, the element which is added first will be removed first.

There are 2 terms used in the queue:

*  front end and rear end.





```python
import queue
q = queue.Queue()
q.put(10)
q.put(20)
q.put(30)
q.put(40)
print("Queue size:", q.qsize())
print("Front element:", q.queue[0])
while not q.empty():
    print("Dequeue:", q.get())
print("Queue size:", q.qsize())
```

**Practice: Implement stack using python**

This program utilizes Python's built-in queue.


Queue class, which implements a FIFO (First In, First Out) queue.


The program enqueues elements into the queue using put(), peeks at the front element using queue[0], dequeues elements using get(), and checks the size of the queue using qsize()
